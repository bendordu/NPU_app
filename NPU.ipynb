```python
ind = 0
diii = []

while ind < raw["EMG"].count():

    first = raw["EMG"].iloc[ind:(ind + 27)].tolist()

    n = 0
    dif = 0

    while n < len(first)-1:
        j = abs((first[n+1] - first[n]) / 0.1)
        dif += j
        n += 1
        
    diii += [dif]
    ind += 27
    
print(diii)

# от 500 до 1500
```
